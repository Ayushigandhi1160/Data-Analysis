```python
import pandas as pd
#Loading the dataset
file_path = '/content/cities_r2.csv'
cities_df = pd.read_csv(file_path)
#Displaying the first five rows of the dataset
cities_df.head()
```

```python
import matplotlib.pyplot as plt
#Sorting the dataset by total population to get top 10 cities
top_10_cities = cities_df.sort_values(by='population_total', ascending=False).head(10)
#Bar Chart: Top 10 cities by population
plt.figure(figsize=(18, 5))
plt.subplot(1, 3, 1)
plt.barh(top_10_cities['name_of_city'], top_10_cities['population_total'], color='skyblue')
plt.xlabel('Total Population')
plt.ylabel('City')
plt.title('Top 10 Cities by Population')
#Pie Chart: Gender Distribution (total population across all cities)
total_male_population = cities_df['population_male'].sum()
total_female_population = cities_df['population_female'].sum()
plt.subplot(1, 3, 2)
plt.pie([total_male_population, total_female_population], labels=['Male', 'Female'],
        autopct='%1.1f%%', colors=['#ff9999','#66b3ff'], startangle=140)
plt.title('Gender Distribution Across All Cities')
#Scatter Plot: Literacy rate vs Sex ratio
plt.subplot(1, 3, 3)
plt.scatter(cities_df['effective_literacy_rate_total'], cities_df['sex_ratio'], alpha=0.5, color='green')
plt.xlabel('Effective Literacy Rate (%)')
plt.ylabel('Sex Ratio (Females per 1000 Males)')
plt.title('Literacy Rate vs Sex Ratio')
plt.tight_layout()
plt.show()
```
